### The super() function

Use super().__setitem__ to call that method on the
superclass, to let it insert or update the key/value pair.

```python
class LastUpdateOrderDict(OrderedDict):

    def __setitem__(self, key, value):
        super().__setitem__(key, value)
        self.move_to_end(key)
```

### Subclassing Built-In Types Is Tricky

Example 14-1. Our "__setitem__" override is ignored by the "__init__" and
"__update__" methods of the built-in dict

```python
class DoppelDict(dict):

    def __setitem__(self, key, value):
        super().__setitem__(key, [value] * 2)
```

```python
dd = DoppelDict(one=1)
```

```python
dd
```

```python
dd['two'] = 2
```

```python
dd.update(three=3)
```

```python
dd
```

```python
dd['four'] = 4
```
